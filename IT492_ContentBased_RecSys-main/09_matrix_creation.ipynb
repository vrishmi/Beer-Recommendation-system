#### Combining prouct vector DataFrame with product details DataFrame

```python
import fastparquet
import pandas as pd
import numpy as np
prodDF = pd.read_parquet("products.parquet", engine="fastparquet")
prod1587DF = pd.read_parquet("prodSVD/prod1587df.parquet", engine="fastparquet")
```

```python
prod1587DF
```

```python
prodDF.drop('review/text', axis=1, inplace=True)
prodDF
```

```python
prod1587Final = pd.concat([prodDF, prod1587DF], axis=1)
prod1587Final
```

```python
prod1587Final.to_parquet("prod1587Final.parquet", engine="fastparquet")
```
