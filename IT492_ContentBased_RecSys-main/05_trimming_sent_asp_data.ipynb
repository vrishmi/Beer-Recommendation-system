## After parsing all the reviews we are left with 1.39 million unique pairs of sentiments-aspect. Here we trim that set to around 43k pairs using basic filtering based on frequency

* ##### `user_aspects.pickle` & `prod_aspect.pikle` contain key-value pair of (sentiment, aspect) appearing in each user preview, product review
* ##### `user_aspects_trimmed.pickle` contains info about those users who have reviewed at least 10 times 
* ##### `prod_aspects_trimmed.pickle` contains info about those prods which have at least 5 reviews 
* ##### `final_prods_aspects` & `final_users_aspects` contain only those pairs which have occured atleast 50 times in both `user_aspects_trimmed.pickle` & `prod_aspects_trimmed.pickle`

```python
import pickle
with open("aspect_freq/prod_aspects_trimmed.pickle", "rb") as f:
    prod_data = pickle.load(f)
with open("aspect_freq/user_aspects_trimmed.pickle", "rb") as f:
    user_data = pickle.load(f)
```

#### `bigram_unique_50.pickle` contains sentiment-aspect pairs which occured atleast 50 times in the corpus.

```python
with open("aspect_freq/bigram_unique_50.pickle", "rb") as f:
    needed_bigrams = set(pickle.load(f))
```

### Filtering product aspect dictionary to contain pairs with minimum frequency of 50

```python
needed_aspects_prod = {}
for ind in prod_data:
    needed_aspects_prod[ind] = []
    for key in prod_data[ind]["bigrams"].keys():
        if key in needed_bigrams:
            needed_aspects_prod[ind].append(key)
    print(f"[*] Examining {ind}")
```

```python
trimmed_prod = {}
for ind in prod_data:
    trimmed_prod[ind] = {key: prod_data[ind]["bigrams"][key] for key in needed_aspects_prod[ind]}
    print(f"[~] Done {ind}")
```

```python
prod_keys = []
for ind in trimmed_prod:
    prod_keys.extend(list(trimmed_prod[ind].keys()))
print(len(set(prod_keys)), len(needed_bigrams))
del prod_keys
```

```python
with open("aspect_freq/final_prods_aspects.pickle", "wb") as f:
    pickle.dump(trimmed_prod, f)
```

```python
del prod_data
del trimmed_prod
```
